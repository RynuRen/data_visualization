# 시각화 테스트

```python
df_plot = df_2017[(df_2017['역명'] == '이태원') & (df_2017['날짜'] == '2017-10-31')]
df_plot.head(3)
```

```python
df_plot = df_plot.melt(id_vars=['날짜', '역명'], var_name='시간', value_name='하차인원').drop(columns=['날짜', '역명'])
df_plot.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_plot2018 = df_2018[(df_2018['역명'] == '이태원') & (df_2018['날짜'] == '2018-10-31')]
df_plot2019 = df_2019[(df_2019['역명'] == '이태원') & (df_2019['날짜'] == '2019-10-31')]
df_plot2020 = df_2020[(df_2020['역명'] == '이태원') & (df_2020['날짜'] == '2020-10-31')]
df_plot2021 = df_2021[(df_2021['역명'] == '이태원') & (df_2021['날짜'] == '2021-10-31')]
```

```python
df_plot2018 = df_plot2018.melt(id_vars=['날짜', '역명'], var_name='시간', value_name='하차인원').drop(columns=['날짜', '역명'])
df_plot2019 = df_plot2019.melt(id_vars=['날짜', '역명'], var_name='시간', value_name='하차인원').drop(columns=['날짜', '역명'])
df_plot2020 = df_plot2020.melt(id_vars=['날짜', '역명'], var_name='시간', value_name='하차인원').drop(columns=['날짜', '역명'])
df_plot2021 = df_plot2021.melt(id_vars=['날짜', '역명'], var_name='시간', value_name='하차인원').drop(columns=['날짜', '역명'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.lineplot(data=df_plot, x='시간', y='하차인원', label='2017', color='violet')
sns.lineplot(data=df_plot2018, x='시간', y='하차인원', label='2018', color='steelblue')
sns.lineplot(data=df_plot2019, x='시간', y='하차인원', label='2019', color='seagreen')
sns.lineplot(data=df_plot2020, x='시간', y='하차인원', label='2020', color='gold')
sns.lineplot(data=df_plot2021, x='시간', y='하차인원', label='2021', color='orangered')
plt.xticks(rotation=45)
plt.title("할로윈 이태원역 하차인원")
```

```python
import folium
from folium import plugins
```

```python
seoul_loc = (37.566535, 126.9779692)
seoul = folium.Map(seoul_loc, zoom_start=12)
seoul
```

```python
heat2017 = heat2017[heat2017['날짜'] == '2017-10-31']
```

```python
heat2017['계'] = heat2017[['05~06', '06~07', '07~08', '08~09', '09~10', '10~11',
       '11~12', '12~13', '13~14', '14~15', '15~16', '16~17', '17~18', '18~19',
       '19~20', '20~21', '21~22', '22~23', '23~24', '00~01']].sum(axis=1)
heat2017.head(3)
```

```python
seoul.add_child(plugins.HeatMap(zip(heat2017['y'], heat2017['x'], heat2017['계']), radius=20))
```
