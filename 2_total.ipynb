```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2017 = pd.read_csv('data/total/total_2017.csv', low_memory=False)
df_2018 = pd.read_csv('data/total/total_2018.csv', low_memory=False)
df_2019 = pd.read_csv('data/total/total_2019.csv', low_memory=False)
df_2020 = pd.read_csv('data/total/total_2020.csv', low_memory=False)
df_2021 = pd.read_csv('data/total/total_2021.csv', low_memory=False)
```

# 년도별 총 하차승객수

```python
total2017 = df_2017.groupby('년')['하차총승객수'].sum()
total2018 = df_2018.groupby('년')['하차총승객수'].sum()
total2019 = df_2019.groupby('년')['하차총승객수'].sum()
total2020 = df_2020.groupby('년')['하차총승객수'].sum()
total2021 = df_2021.groupby('년')['하차총승객수'].sum()
```

```python
total2021
```

```python
total = pd.concat([total2017, total2018, total2019, total2020, total2021])
```

```python
total = pd.DataFrame(total, columns=['하차총승객수']).reset_index()
```

```python
total
```

```python
sns.barplot(data=total, x='년', y='하차총승객수', palette="hls")
```

```python
tmp = total[(total['년'] == 2020) | (total['년'] == 2021)]
tmp
```

```python
plt.ylim(2720000000, 2820000000)
sns.barplot(data=tmp, x='년', y='하차총승객수', palette="hls")
```

# 월별 유동인구 흐름

```python
df = pd.concat([df_2017, df_2018, df_2019, df_2020, df_2021])
df.tail()
```

```python
month = df.groupby('월')['하차총승객수'].sum()
```

```python
month = month.reset_index()
```

```python
plt.ylim(1230000000, 1500000000)
sns.barplot(data=month, x='월', y='하차총승객수', palette="hls")
```

```python
month2017 = df[df['년'] == 2017].groupby('월')['하차총승객수'].sum()

month2017 = month2017.reset_index()
```

```python
plt.ylim(180000000, 340000000)
plt.title('2017년 하차총승객수')
sns.barplot(data=month2017, x='월', y='하차총승객수', palette="hls")
```

```python
month2018 = df[df['년'] == 2018].groupby('월')['하차총승객수'].sum()

month2018 = month2018.reset_index()
```

```python
plt.ylim(180000000, 340000000)
plt.title('2018년 하차총승객수')
sns.barplot(data=month2018, x='월', y='하차총승객수', palette="hls")
```

```python
month2019 = df[df['년'] == 2019].groupby('월')['하차총승객수'].sum()

month2019 = month2019.reset_index()
```

```python
plt.ylim(180000000, 340000000)
plt.title('2019년 하차총승객수')
sns.barplot(data=month2019, x='월', y='하차총승객수', palette="hls")
```

```python
month2020 = df[df['년'] == 2020].groupby('월')['하차총승객수'].sum()

month2020 = month2020.reset_index()
```

```python
plt.ylim(180000000, 340000000)
plt.title('2020년 하차총승객수')
sns.barplot(data=month2020, x='월', y='하차총승객수', palette="hls")
```

```python
month2021 = df[df['년'] == 2021].groupby('월')['하차총승객수'].sum()

month2021 = month2021.reset_index()
```

```python
plt.ylim(180000000, 340000000)
plt.title('2021년 하차총승객수')
sns.barplot(data=month2021, x='월', y='하차총승객수', palette="hls")
```

```python
fig = plt.figure(figsize=(20, 5))
plt.title('연도별 월별 하차총승객수')
sns.barplot(data=df, x='년', y='하차총승객수', hue='월', palette="hls")
plt.legend('', frameon=False)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.lineplot(data=month2017, x='월', y='하차총승객수', label='2017', color='violet')
sns.lineplot(data=month2018, x='월', y='하차총승객수', label='2018', color='steelblue')
sns.lineplot(data=month2019, x='월', y='하차총승객수', label='2019', color='seagreen')
sns.lineplot(data=month2020, x='월', y='하차총승객수', label='2020', color='gold')
sns.lineplot(data=month2021, x='월', y='하차총승객수', label='2021', color='orangered')
plt.xticks(rotation=45)
plt.legend('', frameon=False)
plt.title("17~21년 하차총승객수")
```
