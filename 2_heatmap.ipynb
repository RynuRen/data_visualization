```python
import pandas as pd
import folium
from folium import plugins
```

# 2017-12-22 서울 전체 heatmap

```python
total_2017 = pd.read_csv('data/total/total_2017.csv')
```

```python
total_2017.head(3)
```

```python
target_date = 20171222
```

```python
heat2017 = total_2017[total_2017['사용일자'] == target_date]
```

```python
heat2017.head(3)
```

```python
heat2017 = heat2017[['하차총승객수', 'X좌표', 'Y좌표']]
```

```python
seoul_loc = (37.566535, 126.9779692)
seoul = folium.Map(seoul_loc, zoom_start=12, tiles='Stamen Terrain')
seoul.add_child(plugins.HeatMap(zip(heat2017['Y좌표'], heat2017['X좌표'], heat2017['하차총승객수']), name='2017-12-22', min_opacity=0.1, radius=20))
```

```python
seoul.save('data/total/heatmap/2017_12_22_heatmap.html')
```
