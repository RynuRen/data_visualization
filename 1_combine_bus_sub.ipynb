```python
import pandas as pd
```

```python
data_bus = pd.read_csv('data/bus_total/2017버스.csv')
data_bus.head(3)
```

```python
data_bus.rename(columns={'버스정류장ARS번호':'식별'}, inplace=True)
```

```python
data_bus['구분'] = '버스'
```

```python
data_sub = pd.read_csv('data/subway_total/2017년_지하철데이터.csv')
data_sub.head(3)
```

```python
data_sub.rename(columns={'역명':'식별'}, inplace=True)
```

```python
data_sub['구분'] = '지하철'
```

```python
data = pd.concat([data_bus, data_sub])
data.head(3)
```

```python
data.tail(3)
```

```python
data.info()
```

```python
data.to_csv('data/total/total_2017.csv', index=False)
```
